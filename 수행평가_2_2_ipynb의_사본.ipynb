```python
#데이터 관리 및 시각화에 필요한 모듈들을 불러옵니다.
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.cm as cm
#1.DataFrame()함수를 이용하여 데이터 프레임을 생성해줍니다.
df = pd.DataFrame([[1,2,90,70,85,'kim'],[1,2,65,70,60,'kim'],[2,2,87,75,80,'park']], index = ['lee','koo','ryu'], columns = ['class', 'grade', 'eng', 'math', 'kor', 'teacher'])
df
```

```python
#2.df 데이터프레임의 eng, math, kor 칼럼에만 접근해보세요.
df[['eng','math','kor']]
```

```python
#3.학생 2명(kim,lim)의 데이터를 df에 추가하여 아래와 같이 출력되도록 코드를 작성하세요.
df.loc['kim'] = [2,2,80,80,90,'jung']
df.loc['lim'] = [3,2,100,85,90,'choi']
```

```python
df
```

```python
#4.학생 lee와 koo의 학년(grade)를 1학년으로 변경하세요.
df.loc['lee','grade'] = 1
df.loc['koo','grade'] = 1
df
```

```python
#5.class 열을 삭제하세요.
df.drop(['class'], axis = 1, inplace = True)
df
```

```python
#(참고) values함수와 tolist()를 사용하여 eng열의 데이터들을 불러와 eng_list 리스트를 생성하세요.
eng_list = df['eng'].values.tolist()
#6-1.kor열의 데이터들을 불러와 kor_list 리스트를 생성하세요.
kor_list =df['kor'].values.tolist()
#6-2.math열의 데이터들을 불러와 math_list 리스트를 생성하세요.
math_list = df['math'].values.tolist()

print(eng_list, kor_list, math_list)
```

```python
#7.eng_list와 kor_list를 각각 x, y값으로 하는 산점도를 생성해보세요.
#단 마커의 모양을 마름모꼴로 변경합니다.
x = eng_list
y = kor_list
plt.scatter(x,y,marker='D')
plt.show()
```

```python
#8.name 리스트를 x축, math_list 리스트를 y축으로 하는 막대그래프를 생성해보세요.
#막대그래프의 색상을 지정해보세요. 변경할 때에는 막대그래프를 생성할 때 color 옵션을사용합니다.
name = df.index.tolist()
y = math_list
plt.bar(name,y, color=['g', 'yellow', 'b', 'pink', 'orange'])
plt.show()
```

```python
#9. kor_list값을 이용한 히스토그램을 생성해보세요. 계급 개수는 default값으로 변경하지 않습니다.
plt.hist(kor_list)
plt.show()
```

```python
#10.eng_list, math_list, kor_list를 합친 3행 5열의 2차원 배열 arr을 생성한 뒤 히트맵을 그려봅니다.
#cmap색상을 원하는 것으로 변경하세요.(예시는 Greens를 사용하였음.)
arr = [eng_list, math_list, kor_list]
plt.matshow(arr, cmap=cm.Greens)
plt.show()
```
